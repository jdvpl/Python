```
def consultar(s): 
  return s 
  
def consignar(valor, s): 
  s = s + valor # s += valor 
  return s 
  
def retiro(valor, s): 
  s -= valor # s = s - valor 
  return s 

saldo = 1000000 
bandera = True

while(bandera): 
  print("Por favor escoja una de las siguientes opciones: ") 
  print("1 - para consultar saldo") 
  print("2 - para consignar") 
  print("3 - para retirar") 
  print("0 - para salir") 
  opcion = int(input("\nPor favor digite la opción: ")) 
  if opcion == 1: 
    print(consultar(saldo)) 
  elif opcion == 2: 
    valor = float(input("Por favor ingrese el valor a consignar: ")) 
    print("Despues de la consignación, el saldo de su cuenta es:", end = " ") 
    saldo = consignar(valor, saldo) 
    print(saldo) 
  elif opcion == 3: 
    valor = float(input("Por favor ingrese el valor a retirar: ")) 
    print("Despues del retiro, el saldo de su cuenta es:", end = " ") 
    saldo = retiro(valor, saldo) 
    print(saldo) 
  elif opcion == 0: 
    print("Muchas gracias por utilizar nuestros servicios") 
    bandera = False 
  else: print("Opcion no valida")
```
